## Merge Data Frame

Menggabungkan dua atau lebih dataframe berdasarkan key dimana key nya merupakan kolom/fitur/variabel

### Inner Join

```python
pd.merge(
    employee_dataset, # dataframe 1
    department_dataset, # dataframe 2
    on= 'ID'
)
```

```python
employee_with_department_df = pd.merge(
    employee_dataset, # dataframe 1
    department_dataset, # dataframe 2
    on= 'ID',
    how= 'inner'
)

employee_with_department_df
```

```python
employee_with_department_df1 = pd.merge(
    employee_dataset, # dataframe 1
    department_dataset[['ID', 'Dept_name']], # dataframe 2
    on= 'ID',
    how= 'inner'
)

employee_with_department_df1
```

### Left Join

```python
employee_with_department_df = pd.merge(
    employee_dataset, # dataframe 1
    department_dataset, # dataframe 2
    on= 'ID',
    how= 'left'
)

employee_with_department_df
```

### Right Join

```python
employee_with_department_df = pd.merge(
    employee_dataset, # dataframe 1
    department_dataset, # dataframe 2
    on= 'ID',
    how= 'right'
)

employee_with_department_df
```

## Concatenate Data Frame

Cara kerja .concat() :
* kalau kolom di antara tabel/dataframe yang di concat sama, maka menambahkan sebagai baris baru
* kalau kolom di antara tabel/dataframe yang di concat berbeda, maka menambahkan sebagai kolom baru

```python
dataset1 = pd.DataFrame({
    'Jakarta': [6000, 8000, 8500],
    'Bandung': [6500, 7000, 9000],
    'Tangerang': [5000, 5500, 6000]
}, index= [0, 1,2])

dataset2 = pd.DataFrame({
    'Jakarta': [7000, 8500, 10000],
    'Bandung': [16500, 9000, 2000],
    'Tangerang': [3000, 4500, 4000]
}, index= [3, 4, 5])

dataset3 = pd.DataFrame({
    'Yogyakarta': [2000, 2500, 5000],
    'Semarang': [4500, 5500, 6500],
    'Solo': [5000, 6000, 9500]
}, index = [6, 7 ,8])

display(dataset1, dataset2, dataset3)
```

```python
pd.concat([dataset1, dataset2]) # by default menambahkan sebagai baris baru
```

```python
pd.concat(
    [dataset1, dataset2], 
    axis= 0 # axis = 0 -> menambahkan sebagai baris baru
)
```

```python
pd.concat(
    [dataset1, dataset2], 
    axis= 1 # axis = 1 -> menambahkan sebagai kolom baru
)
```

```python
pd.concat([dataset1, dataset3]) # by default menambahkan sebagai baris baru
```

```python
pd.concat(
    [dataset1, dataset3],
    axis = 0
)
```

```python
pd.concat([dataset1, dataset2, dataset3])
```

```python
benefits_per_year = salaries_dataset_clean.groupby(by = "Year")['Benefits'].mean().reset_index()
benefits_per_year.rename(columns={
    'Year' : 'Tahun',
    'Benefits' : 'Rata-Rata Tunjangan'
}, inplace= True)

benefits_per_year
```

```python
def roundNumber(number):
    return np.round(number, 2)
```

```python
benefits_per_year['Rata-Rata Tunjangan'] = benefits_per_year['Rata-Rata Tunjangan'].apply(roundNumber)
benefits_per_year
```

```python
benefits_per_year['Rata-Rata Tunjangan'] = benefits_per_year['Rata-Rata Tunjangan'].apply(lambda number: np.round(number, 2))
benefits_per_year
```
